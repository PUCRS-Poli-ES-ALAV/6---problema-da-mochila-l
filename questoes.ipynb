   FIBO-REC (n)
      se n ≤ 1
      então devolva n
      senão a ← FIBO-REC (n − 1)
            b ← FIBO-REC (n − 2)
            devolva a + b


````python
def FIBO (n):
    if n <= 1:
      return n 
    else: 
        a = FIBO(n - 1)
        b = FIBO (n - 2)
        return (a + b)
    

print(FIBO(4))
print(FIBO(8))
print(FIBO(16))
print(FIBO(32))

print(FIBO(128))
print(FIBO(1000))
print(FIBO(10000))
````

````python

def fibo(n):
    f = [0] * (n + 1)
    
    f[0] = 0
    if n > 0:
        f[1] = 1

    i = 2
    while i <= n:
        f[i] = f[i - 1] + f[i - 2]
        i += 1

    return f[n]


print(fibo(4))
print(fibo(8))
print(fibo(16))
print(fibo(32))

print(fibo(128))
print(fibo(1000))
print(fibo(10000))
````


````python
import pandas as pd


def memoized_fibo(n):
    f = [0] * (n + 1)
    i = 0
    while i <= n:
        f[i] = -1
        i += 1
    return lookup_fibo(f, n)

def lookup_fibo(f, n):
    if f[n] >= 0:
        return f[n]
    if n <= 1:
        f[n] = n
    else:
        f[n] = lookup_fibo(f, n - 1) + lookup_fibo(f, n - 2)
    return f[n]

def fibo(n):
    f = [0] * (n + 1)

    f[0] = 0
    if n > 0:
        f[1] = 1

    i = 2
    while i <= n:
        f[i] = f[i - 1] + f[i - 2]
        i += 1

    return f[n]

def FIBO(n):
    if n <= 1:
        return n
    else:
        a = FIBO(n - 1)
        b = FIBO(n - 2)
        return a + b


valores_teste = [128, 1000, 10000]

algoritmos = [
    ("memoized_fibo", memoized_fibo),
    ("fibo", fibo),
    ("FIBO", FIBO)
]

dados = []

for nome, funcao in algoritmos:
    for n in valores_teste:
        resultado = funcao(n)
        dados.append({
            "algoritmo": nome,
            "n": n,
            "resultado": resultado
        })

df = pd.DataFrame(dados)

df.to_csv("resultados_fibonacci.csv", index=False, encoding="utf-8")

print(df)
print("\nCSV gerado com sucesso: resultados_fibonacci.csv")
````
